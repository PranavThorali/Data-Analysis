```python
### Extracting Messi's Goals
messi_data = df[df['Player Names'].str.lower() == 'lionel messi']
messi_goals = messi_data[['Year', 'Goals']]
```

```python
### Let's first visualize Messi's Goals
plt.figure(figsize=(10, 6))
sns.lineplot(data=messi_goals, x='Year', y='Goals', markers='o', color='blue' )
plt.title("Lionel Messi's goals over the years")
plt.xlabel("Year")
plt.ylabel("Goals")
plt.show()
```

```python
### now let's look at his xG
messi_goals_xg = messi_data[['Year', 'Goals', 'xG']]

## let's plot it as well
sns.lineplot(data=messi_goals_xg, x='Year', y='Goals', marker='o', color='green')
sns.lineplot(data=messi_goals_xg, x='Year', y='xG', marker='o', color='red')
plt.title("Messi's Goals vs ExpectedGoals (xG)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
### now let's also look at other players such as Luis Suarez and Antoinne Griezmann

antoine_data = df[df['Player Names'].str.lower() == 'antoinne griezmann']
antoine_goals = antoine_data[['Year', 'Goals', 'xG']]

suarez_data = df[df['Player Names'].str.lower() == 'luis suarez']
suarez_goals = suarez_data[['Year', 'Goals', 'xG']]
```

```python
### lets plot them now
plt.figure()
sns.lineplot(data=messi_goals_xg, x='Year', y='Goals', marker='o', color='green')
sns.lineplot(data=messi_goals_xg, x='Year', y='xG', marker='o', color='red')
sns.lineplot(data=antoine_goals, x='Year', y='Goals', marker='x', color='blue')
sns.lineplot(data=antoine_goals, x='Year', y='xG', marker='x', color='black')
sns.lineplot(data=suarez_goals, x='Year', y='Goals', marker='o', color='purple')
sns.lineplot(data=suarez_goals, x='Year', y='xG', marker='o', color='pink')
plt.title("Messi, Suarez and Griezmann Goals and xG plotted ")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
### Now we can try finding the top 10 goal scorers
top_scorers = df.groupby('Player Names')['Goals'].sum().sort_values(ascending=False).head(10)
top_scorers.plot(kind='bar', title="Top 10 Goal Scorers", figsize=(10,5))
```

```python
### let's check the average league goals by league
league_avg_goals = df.groupby('League')['Goals'].mean().sort_values(ascending=False)
league_avg_goals.plot(kind='barh', title='Average Goals per Player by league', figsize=(8,5))
```

```python
### Distribution of Goals and xG
sns.histplot(df['Goals'], bins=20, kde=True)
plt.title('Distribution of Goals')

sns.histplot(df['xG'], bins=20, kde=True)
plt.title('Distribution of Goals and xG')
```

```python
### Average League Goals per year
avg_goals_per_league_year = df.groupby(['Year', 'League'])['Goals'].mean().reset_index()
pivot_avg_goals = avg_goals_per_league_year.pivot(index='Year', columns='League', values='Goals')
```
